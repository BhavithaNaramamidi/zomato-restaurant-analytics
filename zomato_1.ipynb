```python
def cost_bucket(cost):
    if pd.isna(cost):
        return np.nan
    elif cost <= 300:
        return "low"
    elif cost <= 700:
        return "medium"
    else:
        return "high"

df_req["cost_category"] = df_req["approx_cost_for_two"].apply(cost_bucket)
```

```python
def clean_text(text):
    if pd.isna(text):
        return np.nan
    text = str(text).lower()
    text = re.sub(r"[^a-zA-Z0-9, ]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text
```

```python
text_cols = [
    "rest_type",
    "listed_in_type",
    "listed_in_city",
    "cuisines"
]

for col in text_cols:
    df_req[col] = df_req[col].apply(clean_text)
```

```python
df_req["has_dish_liked"] = df_req["dish_liked"].notna().astype(int)
```

```python
def clean_reviews(text):
    if pd.isna(text):
        return np.nan

    text = str(text).lower()

    # 🔹 1. REMOVE escaped & real newlines FIRST
    text = text.replace("\\n", " ").replace("\n", " ")

    # 🔹 2. REMOVE rating tags like 'rated 4.5'
    text = re.sub(r"rated\s*\d+(\.\d+)?", " ", text)

    # 🔹 3. REMOVE the word 'rated' itself
    text = text.replace("rated", " ")

    # 🔹 4. REMOVE tuple / list symbols
    text = re.sub(r"[\[\]\(\)\'\",]", " ", text)

    # 🔹 5. REMOVE non-ASCII safely (without creating junk)
    text = re.sub(r"[^\x00-\x7F]+", " ", text)

    # 🔹 6. KEEP only alphabets and spaces
    text = re.sub(r"[^a-z ]", " ", text)

    # 🔹 7. REMOVE standalone letters like 'n', 'x'
    text = re.sub(r"\b[a-z]\b", " ", text)

    # 🔹 8. NORMALIZE spaces
    text = re.sub(r"\s+", " ", text).strip()

    # 🔹 9. DROP very short junk reviews
    if len(text) < 30:
        return np.nan

    return text
```

```python
df_req["clean_reviews"] = df_req["reviews_list"].apply(clean_reviews)
```

```python
import nltk
from textblob import TextBlob
```

```python
# 1️⃣ Sentiment score function (TextBlob)
def get_sentiment_score(text):
    if pd.isna(text) or str(text).strip() == "":
        return np.nan
    return TextBlob(str(text)).sentiment.polarity
```

```python
# 2️⃣ Limit review length (performance optimization)
df_req["clean_reviews"] = (
    df_req["clean_reviews"]
    .astype(str)
    .str.slice(0, 500)
)
```

```python

# 3️⃣ Apply sentiment scoring
df_req["sentiment_score"] = df_req["clean_reviews"].apply(get_sentiment_score)
```

```python
# 4️⃣ Aggregate to restaurant level
sentiment_agg = (
    df_req.groupby("name", as_index=False)
    .agg(
        sentiment_score=("sentiment_score", "mean")
    )
)
```

```python
# 5️⃣ Handle restaurants with no reviews
sentiment_agg["sentiment_score"] = sentiment_agg["sentiment_score"].fillna(0)
```

```python
def restaurant_sentiment(score):
    if score >= 0.05:
        return "positive"
    elif score <= -0.05:
        return "negative"
    else:
        return "neutral"

sentiment_agg["sentiment"] = sentiment_agg["sentiment_score"].apply(restaurant_sentiment)
```
